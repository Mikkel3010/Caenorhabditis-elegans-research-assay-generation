# STEP 1 : GO ANNOTATIONS EXTRACTION FROM AGR


```python
import pandas as pd
import obonet
gaf_path = r"downloaded_data_ontology_related\wb.gaf"  # WormBase database of biology (wb), https://current.geneontology.org/products/pages/downloads.html

gaf_cols = [
    "DB",
    "DB_Object_ID",
    "DB_Object_Symbol",
    "Qualifier",
    "GO_ID",
    "DB_Reference",
    "Evidence_Code",
    "With_or_From",
    "Aspect",
    "DB_Object_Name",
    "DB_Object_Synonym",
    "DB_Object_Type",
    "Taxon",
    "Date",
    "Assigned_By",
    "Annotation_Extension",
    "Gene_Product_Form_ID",
]

AGR_GO_ANNOTATION_df = pd.read_csv(
    gaf_path,
    sep="\t",
    comment="!",
    header=None,
    names=gaf_cols,
)

AGR_GO_ANNOTATION_df= AGR_GO_ANNOTATION_df.rename(columns={"DB_Object_Symbol":"Gene Symbol"})

AGR_GO_ANNOTATION_df
```

# STEP 2: Full Gene ontology

```python
full_ontology_schema_path = r"downloaded_data_ontology_related\gene_ontology_schema.obo" # https://obofoundry.org/ontology/go.html
 # https://obofoundry.org/ontology/go.html

graph = obonet.read_obo(full_ontology_schema_path)


#   Example data
#  {'GO:0000001': 'mitochondrion inheritance',
#  'GO:0000006': 'high-affinity zinc transmembrane transporter activity',
#  'GO:0000007': 'low-affinity zinc ion transmembrane transporter activity',
#  'GO:0000009': 'alpha-1,6-mannosyltransferase activity',
go_name_map = {
    go_id: data.get("name")
    for go_id, data in graph.nodes(data=True)
}
go_namespace_map = { #name space is 3 types: Root terms for the wb data: either molecular_function, biological_process or cellular_component
    go_id: data.get("namespace")
    for go_id, data in graph.nodes(data=True)
}

# MAP to existing GO_ID in the populated AGR data.
AGR_GO_ANNOTATION_df["GO_DESCRIPTION"] = AGR_GO_ANNOTATION_df["GO_ID"].map(go_name_map)
AGR_GO_ANNOTATION_df["GO_NAMESPACE"] = AGR_GO_ANNOTATION_df["GO_ID"].map(go_namespace_map)
AGR_GO_ANNOTATION_df
```

# STEP 3: Simplicfication list based on ocurrence

```python
import pandas as pd


def send_upstream(go_id, graph, amount=1.0):

    upstream_bucket = {}

    def visit(term_id, weight):
        upstream_bucket[term_id] = upstream_bucket.get(term_id, 0.0) + weight

        parents = list(graph.successors(term_id))  # obonet: child -> parent

        if not parents:
            return

        split_weight = weight / len(parents)

        for parent_id in parents:
            visit(parent_id, split_weight)

    if pd.notna(go_id) and str(go_id) in graph:
        visit(str(go_id), amount)

    return upstream_bucket


def build_upstream_support(df, graph, go_id_col="GO_ID"):

    accumulation_bucket = {}

    go_counts = (
        df[go_id_col]
        .dropna()
        .astype(str)
        .value_counts()
        .to_dict()
    )

    for go_id, count in go_counts.items():
        upstream_bucket = send_upstream(
            go_id=go_id,
            graph=graph,
            amount=count,
        )

        for upstream_id, support in upstream_bucket.items():
            accumulation_bucket[upstream_id] = (
                accumulation_bucket.get(upstream_id, 0.0) + support
            )

    return accumulation_bucket


def upstream_support_to_df(upstream_support, graph):
    """
    Convert upstream support bucket into a readable dataframe.
    """
    rows = []

    for go_id, support in upstream_support.items():
        rows.append({
            "GO_ID": go_id,
            "GO_DESCRIPTION": graph.nodes[go_id].get("name", go_id)
            if go_id in graph else go_id,
            "GO_namespace": graph.nodes[go_id].get("namespace", pd.NA)
            if go_id in graph else pd.NA,
            "support_weight": support,
        })

    return (
        pd.DataFrame(rows)
        .sort_values("support_weight", ascending=False)
        .reset_index(drop=True)
    )

upstream_support = build_upstream_support(
    AGR_GO_ANNOTATION_df,
    graph=graph,
    go_id_col="GO_ID",
)

upstream_support_df = upstream_support_to_df(
    upstream_support,
    graph=graph,
)


upstream_support_df = upstream_support_df[3:59].sort_values(
    ["support_weight"], ascending=[False]) # same as AGR sizewise #removes the 3 roots

upstream_support_df = upstream_support_df.sort_values(
    ["GO_namespace", "support_weight"],
    ascending=[True, False],
).reset_index(drop=True)

upstream_support_df
```
